```python
# download tracking results 
s3.download_file(Bucket=bucket, Key='nfl-data/tracking_results_2020-07-2517-53-51.079650.json', Filename='tracking_results.json')
```

```python
# load results
with open('tracking_results.json','r') as f:
    result = json.load(f)
```
